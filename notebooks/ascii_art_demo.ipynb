````python
from PIL import Image   
import numpy as np      

ASCII_CHARS = "@%#*+=-:. "  

def resize_image(image, new_width=100):
    width, height = image.size
    aspect_ratio = height / width
    new_height = int(aspect_ratio * new_width * 0.55) 
    return image.resize((new_width, new_height))

def grayify(image):
    return image.convert("L")

def pixels_to_ascii(image):
    pixels = np.array(image)
    ascii_str = ""
    for row in pixels:
        for pixel in row:
            ascii_str += ASCII_CHARS[pixel // 25] 
        ascii_str += "\n" 
    return ascii_str

def image_to_ascii(image_path, new_width=100):
    image = Image.open(image_path)
    image = resize_image(image, new_width)
    image = grayify(image)
    ascii_art = pixels_to_ascii(image)
    return ascii_art

ascii_output = image_to_ascii("your_image.jpg", new_width=80)
print(ascii_output)
````

````python
ascii_output = image_to_ascii("cat.jpg", new_width=80)
print(ascii_output)
````

````python
with open("ascii_output.txt", "w") as f:
    f.write(ascii_output)

print("ASCII art saved to ascii_output.txt")
````
